**–ó–∞–¥–∞–Ω–∏–µ 5. –í –¥–≤—É—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö –µ—Å—Ç—å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è. –û–¥–∏–Ω –∏–∑ –Ω–∏—Ö ‚Äî `days_employed`. –ü—Ä–æ–ø—É—Å–∫–∏ –≤ —ç—Ç–æ–º —Å—Ç–æ–ª–±—Ü–µ –≤—ã –æ–±—Ä–∞–±–æ—Ç–∞–µ—Ç–µ –Ω–∞ —Å–ª–µ–¥—É—é—â–µ–º —ç—Ç–∞–ø–µ. –î—Ä—É–≥–æ–π —Å—Ç–æ–ª–±–µ—Ü —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ ‚Äî `total_income` ‚Äî —Ö—Ä–∞–Ω–∏—Ç –¥–∞–Ω–Ω—ã–µ –æ –¥–æ—Ö–æ–¥–∞—Ö. –ù–∞ —Å—É–º–º—É –¥–æ—Ö–æ–¥–∞ —Å–∏–ª—å–Ω–µ–µ –≤—Å–µ–≥–æ –≤–ª–∏—è–µ—Ç —Ç–∏–ø –∑–∞–Ω—è—Ç–æ—Å—Ç–∏, –ø–æ—ç—Ç–æ–º—É –∑–∞–ø–æ–ª–Ω–∏—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —ç—Ç–æ–º —Å—Ç–æ–ª–±—Ü–µ –Ω—É–∂–Ω–æ –º–µ–¥–∏–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø–æ –∫–∞–∂–¥–æ–º—É —Ç–∏–ø—É –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ `income_type`. –ù–∞–ø—Ä–∏–º–µ—Ä, —É —á–µ–ª–æ–≤–µ–∫–∞ —Å —Ç–∏–ø–æ–º –∑–∞–Ω—è—Ç–æ—Å—Ç–∏ `—Å–æ—Ç—Ä—É–¥–Ω–∏–∫` –ø—Ä–æ–ø—É—Å–∫ –≤ —Å—Ç–æ–ª–±—Ü–µ `total_income` –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –∑–∞–ø–æ–ª–Ω–µ–Ω –º–µ–¥–∏–∞–Ω–Ω—ã–º –¥–æ—Ö–æ–¥–æ–º —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö –∑–∞–ø–∏—Å–µ–π —Å —Ç–µ–º –∂–µ —Ç–∏–ø–æ–º.**

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

### –û–±—Ä–∞–±–æ—Ç–∫–∞ –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π

**–ó–∞–¥–∞–Ω–∏–µ 6. –í –¥–∞–Ω–Ω—ã—Ö –º–æ–≥—É—Ç –≤—Å—Ç—Ä–µ—á–∞—Ç—å—Å—è –∞—Ä—Ç–µ—Ñ–∞–∫—Ç—ã (–∞–Ω–æ–º–∞–ª–∏–∏) ‚Äî –∑–Ω–∞—á–µ–Ω–∏—è, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –æ—Ç—Ä–∞–∂–∞—é—Ç –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∏ –ø–æ—è–≤–∏–ª–∏—Å—å –ø–æ –∫–∞–∫–æ–π-—Ç–æ –æ—à–∏–±–∫–µ. —Ç–∞–∫–∏–º –∞—Ä—Ç–µ—Ñ–∞–∫—Ç–æ–º –±—É–¥–µ—Ç –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π —Ç—Ä—É–¥–æ–≤–æ–≥–æ —Å—Ç–∞–∂–∞ –≤ —Å—Ç–æ–ª–±—Ü–µ `days_employed`. –î–ª—è —Ä–µ–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —ç—Ç–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ. –û–±—Ä–∞–±–æ—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —ç—Ç–æ–º —Å—Ç–æ–ª–±—Ü–µ: –∑–∞–º–µ–Ω–∏—Ç–µ –≤—Å–µ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º–∏ —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ `abs()`.**

```python
data['days_employed'] = data['days_employed'].abs()
```

**–ó–∞–¥–∞–Ω–∏–µ 7. –î–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∏–ø–∞ –∑–∞–Ω—è—Ç–æ—Å—Ç–∏ –≤—ã–≤–µ–¥–∏—Ç–µ –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ç—Ä—É–¥–æ–≤–æ–≥–æ —Å—Ç–∞–∂–∞ `days_employed` –≤ –¥–Ω—è—Ö.**

**–ó–∞–¥–∞–Ω–∏–µ 8. –í—ã–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–µ—á–µ–Ω—å —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ `children`.**

**–ó–∞–¥–∞–Ω–∏–µ 9. –í —Å—Ç–æ–ª–±—Ü–µ `children` –µ—Å—Ç—å –¥–≤–∞ –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è. –£–¥–∞–ª–∏—Ç–µ —Å—Ç—Ä–æ–∫–∏, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è —Ç–∞–∫–∏–µ –∞–Ω–æ–º–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ `data`.**

**–ó–∞–¥–∞–Ω–∏–µ 10. –ï—â—ë —Ä–∞–∑ –≤—ã–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–µ—á–µ–Ω—å —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ `children`, —á—Ç–æ–±—ã —É–±–µ–¥–∏—Ç—å—Å—è, —á—Ç–æ –∞—Ä—Ç–µ—Ñ–∞–∫—Ç—ã —É–¥–∞–ª–µ–Ω—ã.**

### –£–¥–∞–ª–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ (–ø—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ)

**–ó–∞–¥–∞–Ω–∏–µ 11. –ó–∞–ø–æ–ª–Ω–∏—Ç–µ –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ `days_employed` –º–µ–¥–∏–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –ø–æ –∫–∞–∂–¥–æ–≥–æ —Ç–∏–ø–∞ –∑–∞–Ω—è—Ç–æ—Å—Ç–∏ `income_type`.**

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['days_employed'].isna()), 'days_employed'] = \
    data.loc[(data['income_type'] == t), 'days_employed'].median()
```

**–ó–∞–¥–∞–Ω–∏–µ 12. –£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ –≤—Å–µ –ø—Ä–æ–ø—É—Å–∫–∏ –∑–∞–ø–æ–ª–Ω–µ–Ω—ã. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ —Å–µ–±—è –∏ –µ—â—ë —Ä–∞–∑ –≤—ã–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ —Å –ø–æ–º–æ—â—å—é –¥–≤—É—Ö –º–µ—Ç–æ–¥–æ–≤.**

### –ò–∑–º–µ–Ω–µ–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö

### –û–±—Ä–∞–±–æ—Ç–∫–∞ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤

**–ó–∞–¥–∞–Ω–∏–µ 14. –û–±—Ä–∞–±–æ—Ç–∞–π—Ç–µ –Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –≤ —Å—Ç–æ–ª–±—Ü–µ `education`. –í —ç—Ç–æ–º —Å—Ç–æ–ª–±—Ü–µ –µ—Å—Ç—å –æ–¥–Ω–∏ –∏ —Ç–µ –∂–µ –∑–Ω–∞—á–µ–Ω–∏—è, –Ω–æ –∑–∞–ø–∏—Å–∞–Ω–Ω—ã–µ –ø–æ-—Ä–∞–∑–Ω–æ–º—É: —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –∑–∞–≥–ª–∞–≤–Ω—ã—Ö –∏ —Å—Ç—Ä–æ—á–Ω—ã—Ö –±—É–∫–≤. –ü—Ä–∏–≤–µ–¥–∏—Ç–µ –∏—Ö –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É.**

**–ó–∞–¥–∞–Ω–∏–µ 15. –í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫-–¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö. –ï—Å–ª–∏ —Ç–∞–∫–∏–µ —Å—Ç—Ä–æ–∫–∏ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É—é—Ç, —É–¥–∞–ª–∏—Ç–µ –∏—Ö.**

### –ö–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö

**–ó–∞–¥–∞–Ω–∏–µ 16. –ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤, —É–∫–∞–∑–∞–Ω–Ω—ã—Ö –Ω–∏–∂–µ, —Å–æ–∑–¥–∞–π—Ç–µ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ `data` —Å—Ç–æ–ª–±–µ—Ü `total_income_category` —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏:**

- 0‚Äì30000 ‚Äî `'E'`;
- 30001‚Äì50000 ‚Äî `'D'`;
- 50001‚Äì200000 ‚Äî `'C'`;
- 200001‚Äì1000000 ‚Äî `'B'`;
- 1000001 –∏ –≤—ã—à–µ ‚Äî `'A'`.


**–ù–∞–ø—Ä–∏–º–µ—Ä, –∫—Ä–µ–¥–∏—Ç–æ–ø–æ–ª—É—á–∞—Ç–µ–ª—é —Å –¥–æ—Ö–æ–¥–æ–º 25000 –Ω—É–∂–Ω–æ –Ω–∞–∑–Ω–∞—á–∏—Ç—å –∫–∞—Ç–µ–≥–æ—Ä–∏—é `'E'`, –∞ –∫–ª–∏–µ–Ω—Ç—É, –ø–æ–ª—É—á–∞—é—â–µ–º—É 235000, ‚Äî `'B'`. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é —Å –∏–º–µ–Ω–µ–º `categorize_income()` –∏ –º–µ—Ç–æ–¥ `apply()`.**

```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

**–ó–∞–¥–∞–Ω–∏–µ 17. –í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω –ø–µ—Ä–µ—á–µ–Ω—å —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Ü–µ–ª–µ–π –≤–∑—è—Ç–∏—è –∫—Ä–µ–¥–∏—Ç–∞ –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ `purpose`.**

```python
data['purpose'].unique()
```

**–ó–∞–¥–∞–Ω–∏–µ 18. –°–æ–∑–¥–∞–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ `purpose` —Å—Ñ–æ—Ä–º–∏—Ä—É–µ—Ç –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü `purpose_category`, –≤ –∫–æ—Ç–æ—Ä—ã–π –≤–æ–π–¥—É—Ç —Å–ª–µ–¥—É—é—â–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏:**

- `'–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–º'`,
- `'–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å—é'`,
- `'–ø—Ä–æ–≤–µ–¥–µ–Ω–∏–µ —Å–≤–∞–¥—å–±—ã'`,
- `'–ø–æ–ª—É—á–µ–Ω–∏–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è'`.

**–ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ `purpose` –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –ø–æ–¥—Å—Ç—Ä–æ–∫–∞ `'–Ω–∞ –ø–æ–∫—É–ø–∫—É –∞–≤—Ç–æ–º–æ–±–∏–ª—è'`, —Ç–æ –≤ —Å—Ç–æ–ª–±—Ü–µ `purpose_category` –¥–æ–ª–∂–Ω–∞ –ø–æ—è–≤–∏—Ç—å—Å—è —Å—Ç—Ä–æ–∫–∞ `'–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–º'`.**

**–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é —Å –∏–º–µ–Ω–µ–º `categorize_purpose()` –∏ –º–µ—Ç–æ–¥ `apply()`. –ò–∑—É—á–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ –≤ —Å—Ç–æ–ª–±—Ü–µ `purpose` –∏ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, –∫–∞–∫–∏–µ –ø–æ–¥—Å—Ç—Ä–æ–∫–∏ –ø–æ–º–æ–≥—É—Ç –≤–∞–º –ø—Ä–∞–≤–∏–ª—å–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∫–∞—Ç–µ–≥–æ—Ä–∏—é.**

```python
def categorize_purpose(row):
    try:
        if '–∞–≤—Ç–æ–º' in row:
            return '–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–º'
        elif '–∂–∏–ª' in row or '–Ω–µ–¥–≤–∏–∂' in row:
            return '–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å—é'
        elif '—Å–≤–∞–¥' in row:
            return '–ø—Ä–æ–≤–µ–¥–µ–Ω–∏–µ —Å–≤–∞–¥—å–±—ã'
        elif '–æ–±—Ä–∞–∑–æ–≤' in row:
            return '–ø–æ–ª—É—á–µ–Ω–∏–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è'
    except:
        return '–Ω–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏–∏'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

```python
#–°–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—É—é —Ç–∞–±–ª–∏—Ü—É debt_from_children. –ò–∑ —Ç–∞–±–ª–∏—Ü—ã data –ø–µ—Ä–µ–Ω–µ—Å—ë–º —Å—Ç–æ–ª–±—Ü—ã: 'children', 'debt'
debt_from_children = data[['children', 'debt']]
#–ù–∞–π–¥—ë–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–µ–º–µ–π –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–¥–æ–ª–∂–µ–Ω–Ω–æ—Å—Ç–µ–π –ø–æ –∫—Ä–µ–¥–∏—Ç—É –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π
#–°–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ, —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ—Ö—Ä–∞–Ω–∏–º –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π debt_from_children_grouped 
debt_from_children_grouped = debt_from_children.groupby('children').agg({'debt': ['count', 'sum']})
#–ü–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–Ω–≤–µ—Ä—Å–∏—é –ø–æ –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –¥–µ—Ç–µ–π. –†–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ—Ö—Ä–∞–Ω–∏–º –≤ —Å—Ç–æ–ª–±—Ü–µ 'conversion'  
debt_from_children_grouped['conversion'] = (debt_from_children_grouped['debt']['sum'] / debt_from_children_grouped['debt']['count'])*100
print(debt_from_children_grouped)
```

```python
# –í–∞—à –∫–æ–¥ –±—É–¥–µ—Ç –∑–¥–µ—Å—å. –í—ã –º–æ–∂–µ—Ç–µ —Å–æ–∑–¥–∞–≤–∞—Ç—å –Ω–æ–≤—ã–µ —è—á–µ–π–∫–∏.
debt_for_family_status = data[['family_status', 'debt']]
debt_for_family_status_grouped = debt_for_family_status.groupby('family_status').agg({'debt': ['count', 'sum']}) 
debt_for_family_status_grouped['conversion'] = (debt_for_family_status_grouped['debt']['sum'] / debt_for_family_status_grouped['debt']['count'])*100
#–≤—ã–≤–µ–¥–µ–º –¥–∞–Ω–Ω—ã–µ, –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–≤ –ø–æ —Å—Ç–æ–ª–±—Ü—É 'conversion' –≤ –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è
print(debt_for_family_status_grouped.sort_values('conversion', ascending = False))
```
